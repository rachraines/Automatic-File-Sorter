Automatic File Sorter in File Explorer

```python
import os, shutil
path = r"C:/Users/rfros/OneDrive/Documents/MySQL Data Files/Automatic File Sorter Project/"
```

```python
# Shows all files in the path
file_names = os.listdir(path)
```

```python
# Creates file-type folders if not already exists
folder_names = ['csv files', 'image files', 'txt files']
for folder in folder_names:
    if not os.path.exists(path + folder):
        os.makedirs(path + folder)

# Checks each file and sorts into correct folder
for file in file_names:
    if '.csv' in file and not os.path.exists(path + 'csv files/' + file):
        shutil.move(path + file, path + 'csv files/' + file)
    elif '.jpg' in file and not os.path.exists(path + 'image files/' + file):
        shutil.move(path + file, path + 'image files/' + file)    
    elif '.txt' in file and not os.path.exists(path + 'txt files/' + file):
        shutil.move(path + file, path + 'txt files/' + file)   
```
